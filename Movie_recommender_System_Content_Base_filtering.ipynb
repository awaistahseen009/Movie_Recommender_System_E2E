```python
def convert_string_to_list(txt):
    return ast.literal_eval(txt)
```

```python
cols_to_convert=['genres','keywords','cast','crew']
for col in cols_to_convert:
    movies[col]=movies[col].apply(convert_string_to_list)
```

## Now we have successfully converted the strings into list lets fetch the contents from those lists

```python
def fetch_keywords(text_list):
    keywords=[]
    for txt in text_list:
        keywords.append(txt['name'])
    return keywords # We can use same function for genres and keywords
```

```python
len(movies['cast'][0]) # lets check for cast, We cannot fetch and its also not feasible to show whole cast we will just select first 5 actors
```

```python
def fetch_cast(cast_list):
    cast=list()
    for i,cst in enumerate(cast_list):
        if i<3:
            cast.append(cst['name'])
        else:
            break
    return cast
        
```

```python
def fetch_crew(crew_list): # In a movie the most important crew and for people also search is director so we have to fetch him/her.
    crew=list()
    for crw in crew_list:
        if crw['job']=='Director':
            crew.append(crw['name'])
    return crew
```

```python
movies['keywords']=movies['keywords'].apply(fetch_keywords)
movies['genres']=movies['genres'].apply(fetch_keywords)
```

```python
movies['crew']=movies['crew'].apply(fetch_crew)
movies['cast']=movies['cast'].apply(fetch_cast)
```

```python
movies.head()
```

```python
def remove_spaces(text):
    return text.replace(" ","")
```

```python
movies['cast_ws'] = movies['cast'].apply(lambda x: [remove_spaces(x) for x in x])
```

```python
movies['genres_ws'] = movies['genres'].apply(lambda x: [remove_spaces(x) for x in x])
```

```python
movies.head(2)
```

```python
movies['overview']=movies['overview'].apply(lambda x: str(x).split())
```

```python
movies.head(2)
```

```python
movies['tags']=movies['overview'] + movies['genres_ws'] + movies['keywords'] + movies['cast_ws'] + movies['crew']
```

```python
movies['tags']=movies['tags'].apply(lambda x: ' '.join(x))
```

```python
movies['tags'].head(3)
```

```python
movies['tags']=movies['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=5000,stop_words='english')
```

```python
embeddings = tfidf.fit_transform(movies['tags']).toarray()
```

```python
embeddings.shape
```

```python
import difflib
```

```python
all_movies=movies['title'].values
```

```python
all_movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(embeddings)
```

```python
similarity
```

```python
movie_name = "iron man"
corrected_movie_name = difflib.get_close_matches(movie_name, all_movies)
```

```python
corrected_movie_name[0]
```

```python
def recommend_movies(movie_name , n_movies):
    movie_name = difflib.get_close_matches(movie_name, all_movies)[0]
    movie_index=movies[movies['title'] == movie_name].index[0]
    distances = sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])
    for i in distances[:n_movies]:
        print(movies.iloc[i[0]].title)
```

```python
recommend_movies("iron man",10)
```
